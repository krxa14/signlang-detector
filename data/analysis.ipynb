# SignLang EDA
Class distribution, augmentation, and feature importance.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
from collections import Counter
from pathlib import Path
import pandas as pd, matplotlib.pyplot as plt
from data.dataset import DATASET_ROOT, get_class_mapping
cm = get_class_mapping()
lbls=[]
for p in (DATASET_ROOT/'train'/'labels').glob('*.txt'):
    try: lbls.append(int(open(p).readline().split()[0]))
    except: pass
s = pd.Series([cm[i] for i in lbls]).value_counts()
s.head(30).plot(kind='bar', figsize=(14,4)); plt.title('Top-30 class freq'); plt.show()
print('classes w/ <10 samples:', (s<10).sum())
```

```python
# Feature importance from RF (requires training)
from models.rf_model import RFModel, DEFAULT_WEIGHTS
if pathlib.Path(DEFAULT_WEIGHTS).exists():
    m = RFModel(); m.load(); imp = m.get_feature_importance()
    plt.figure(figsize=(14,3)); plt.bar(range(63), imp); plt.title('RF feature importance (63 keypoint dims)'); plt.show()
else:
    print('Train RF first: python scripts/train.py --skip-yolo --skip-cnn')
```
